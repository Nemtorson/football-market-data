```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd
from pathlib import Path  
```

```python
links = ["https://fbref.com/en/comps/Big5/2022-2023/keepers/players/2022-2023-Big-5-European-Leagues-Stats",
         "https://fbref.com/en/comps/Big5/2017-2018/passing/players/2017-2018-Big-5-European-Leagues-Stats",
         "https://fbref.com/en/comps/Big5/2017-2018/shooting/players/2017-2018-Big-5-European-Leagues-Stats",
         "https://fbref.com/en/comps/Big5/2023-2024/defense/players/2023-2024-Big-5-European-Leagues-Stats",
         "https://fbref.com/en/comps/Big5/2017-2018/possession/players/2017-2018-Big-5-European-Leagues-Stats"
         ]
```

```python
download_service = Service()

driver = webdriver.Chrome(service=download_service)

driver.get(links[3])
driver.maximize_window()

driver.execute_script("window.scrollBy(0, 1700);")

time.sleep(2)

#stats_passing_control
#stats_shooting_control
#stats_keeper_control
#stats_defense_control

button_id = "stats_defense_control"

web_button = driver.find_element(by=By.ID, value=button_id)
web_button.click()
```

```python
#stats_passing
#stats_shooting
#stats_keeper
#stats_defense


table = driver.find_element(By.XPATH, '//*[@id="stats_defense"]/tbody')
header_rows = driver.find_element(By.XPATH, '//*[@id="stats_defense"]/thead/tr[2]')
rows = table.find_elements(By.TAG_NAME, 'tr')
cols = table.find_elements(By.TAG_NAME, 'td')
```
